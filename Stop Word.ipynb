```python
import spacy
```

```python
from spacy.lang.en.stop_words import STOP_WORDS
STOP_WORDS
```

```python
nlp=spacy.load("en_core_web_sm")
doc=nlp("We just opened our wings, the flying part is coming soon")

for token in doc:
  if token.is_stop:
      print(token)
```

```python
def preprocess(text):
   doc=nlp(text)
  
   no_stop_word=[token.text for token in doc if not token.is_stop and not token.is_punct]
   return " ".join(no_stop_word)
          
       
```

```python
preprocess("Musk wants time o prepare for a trial over his")
```

```python
preprocess("The other is not other but your devine brother")
```
